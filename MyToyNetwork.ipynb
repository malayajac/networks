```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
MyEdgeList = [(0, 1), (0, 2), (0, 2),
              (1, 2),
              (2, 3),
              (3, 4), (3, 5),
              (4, 5), (4, 6),
              (5, 6)]
```

```python
G = nx.Graph()
```

```python
G.add_edges_from(MyEdgeList)
```

```python
nx.draw_networkx(G)
plt.show()
```
